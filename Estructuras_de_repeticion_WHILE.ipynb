#ESTRUCTURA DE REPETICION MIENTRAS
Es una estructura de repeticion booleana
empleada cuando no se sabe cuantas veces se
debe repetir la condicion
Comprende de
1. Una condicion inicial
2. Un centinela que evalua si se cumple la condicion final
3. Un modificador de condicion inicial

A diferencia de la estructura para, el modificador puede ser un contador, un acumulador o lo que sea.

```python
#EJEMPLO 1: Planificador de compras
#Hacer un planificador de compras que se detenga cuando ya no tenemos dinero
```

```python
ahorro=int(input("Con cuanto dinero cuento?"))
opc=input("Deseo comprar algo?")
while ahorro>=0 and opc=="s":
  precio=int(input("Cuanto cuesta eso?"))
  if precio<=ahorro:
      print("Lo compro!")
      ahorro=ahorro-precio #acumulador
  else:
    print("No me alcanza")

opc=input("Deseo comprar algo?")
```
